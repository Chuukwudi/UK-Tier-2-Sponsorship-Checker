```python
import requests
import json
import pandas as pd

url = "https://hvpfqi4bob.execute-api.eu-west-2.amazonaws.com/sponsorship-checker"

payload = {
            "word": "Norfolk", 
            "based_on": "org", 
            "route": "Skilled Worker"
          }

payload = json.dumps(payload)
headers = {
  'Content-Type': 'text/plain'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)

dd = json.loads(response.text)

dd
```

```python
word = input("Enter the word you want to search here: ")

based_on = input("Enter where you want to search here: (select 1 for organisation, 2 for Town/City, 3 for County)")

route = input("What visa route?: (1 for 'Skilled Worker', 2 for 'Creative Worker', 3 for 'Global Business Mobility: Senior or Specialist Worker', and 4 for 'Tier 2 Ministers of Religion')")


based_on_dict = {"1": "org", "2": "toc", "3": "county"}
route_dict = {"1": "Skilled Worker", "2": "Creative Worker", "3": "Global Business Mobility: Senior or Specialist Worker", "4": "Tier 2 Ministers of Religion"}


url = "https://hvpfqi4bob.execute-api.eu-west-2.amazonaws.com/sponsorship-checker"

payload = {
            "word": word, 
            "based_on": based_on_dict[based_on], 
            "route": route_dict[route]
          }

print(payload)

payload = json.dumps(payload)
headers = {
  'Content-Type': 'text/plain'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)

dd = json.loads(response.text)

dd
```
